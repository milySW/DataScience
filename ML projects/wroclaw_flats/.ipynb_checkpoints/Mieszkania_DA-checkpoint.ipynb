## Podział datasetu na testowy i treningowy

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = my_train_test_split(data,
                                                                       'Price',
                                                                       'apartamentowiec',
                                                                       ['Popularity', 'Time', 'Date', 'Sold_in_30'],
                                                                       True)

X_train_clf, X_test_clf, y_train_clf, y_test_clf = my_train_test_split(data,
                                                                       'Sold_in_30',
                                                                       'apartamentowiec',
                                                                       ['Popularity', 'Time', 'Date'])

indices = np.random.permutation(len(X_test_reg) + len(X_train_reg))

X_reg, y_reg = pd.concat([X_test_reg, X_train_reg]).iloc[indices], pd.concat([y_test_reg, y_train_reg]).iloc[indices]
X_clf, y_clf = pd.concat([X_test_clf, X_train_clf]).iloc[indices], pd.concat([y_test_clf, y_train_clf]).iloc[indices]
```

# Regresja

### Regresja wieloraka

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train_reg,  y_train_reg)

scores = cross_val_score(lin_reg, X_reg, y_reg, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(abs(scores))
np.mean(rmse_scores)
```

```python
lin_reg_area = LinearRegression()

scores = cross_val_score(lin_reg_area,np.array(X_reg[['Area_to_Location_mean']]).reshape(-1, 1), y_reg, scoring="neg_mean_squared_error", cv=10)
rmse_scores_area = np.sqrt(abs(scores))
np.mean(rmse_scores_area)
```

## Multivariate adaptive regression spline

https://github.com/scikit-learn-contrib/py-earth

```python
!pip install git+https://github.com/scikit-learn-contrib/py-earth.git@v0.2dev
```

```python
from pyearth import Earth
from sklearn.ensemble import AdaBoostRegressor

mars = Earth(max_degree=2, penalty=1.5)
mars = AdaBoostRegressor(base_estimator=mars, n_estimators=15, learning_rate=0.1, loss="exponential")

scores = cross_val_score(mars, X_reg, list(y_reg['Price']), scoring="neg_mean_squared_error", cv=5)
rmse_scores = np.sqrt(-scores)
np.mean(rmse_scores)
```

# Klasyfikacja

```python
import pickle

from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import (
    GradientBoostingClassifier, 
    RandomForestClassifier,
    VotingClassifier
)
```

## Regresja logistyczna


```python
models_dir = os.getcwd() + '/gdrive/My Drive/Mieszkania_Wrocław/models'
```

```python
if 'clf_logistic_regression.sav' in os.listdir(models_dir):
  clf_logistic_regression = pickle.load(open(models_dir + '/clf_logistic_regression.sav', 'rb'))
else:
    clf_logistic_regression = prepare_model(LogisticRegression(), 
                                            {'solver':['lbfgs', 'sag', 'saga'], 
                                             'max_iter':[1e4]}, 
                                            X_train_clf, 
                                            y_train_clf.values.ravel(),
                                            'f1')
    
    filename = models_dir + '/clf_logistic_regression.sav'
    pickle.dump(clf_logistic_regression, open(filename, 'wb'))
```

```python
model_tests(clf_logistic_regression, X_test_clf,  y_test_clf.values.ravel())
```

```python
if 'clf_decision_tree.sav' in os.listdir(models_dir):
    clf_decision_tree = pickle.load(open(models_dir + '/clf_decision_tree.sav', 'rb'))
else:
    clf_decision_tree = prepare_model(DecisionTreeClassifier(), 
                                      {'min_samples_split':[20, 390, 1000], 
                                       'max_depth':[None, 5, 10, 20],
                                       'random_state':[12]}, 
                                      X_train_clf, 
                                      y_train_clf.values.ravel(),
                                      'f1')
    
    filename = models_dir + '/clf_decision_tree.sav'
    pickle.dump(clf_decision_tree, open(filename, 'wb'))
```

```python
model_tests(clf_decision_tree, X_test_clf,  y_test_clf.values.ravel())
```

```python
if 'clf_svc.sav' in os.listdir(models_dir):
    clf_svc = pickle.load(open(models_dir + '/clf_svc.sav', 'rb'))
else:
    clf_svc = prepare_model(SVC(), 
                            {'C':[1.0, 2.0], 
                             'class_weight':['balanced'],
                             'random_state':[12],
                             'gamma':['scale'],
                             'kernel':['poly', 'rbf', 'sigmoid']},
                            X_train_clf,
                            y_train_clf.values.ravel(),
                            'f1')
    
    filename = models_dir + '/clf_svc.sav'
    pickle.dump(clf_svc, open(filename, 'wb'))
```

```python
model_tests(clf_svc, X_test_clf,  y_test_clf.values.ravel())
```

```python
if 'clf_sgd.sav' in os.listdir(models_dir):
    clf_sgd = pickle.load(open(models_dir + '/clf_sgd.sav', 'rb'))
else:
    clf_sgd = prepare_model(SGDClassifier(), 
                            {'penalty':['l1', 'l2', 'elasticnet'],
                             'alpha':[0.0001, 0.01],
                             'random_state':[12]},
                            X_train_clf,
                            y_train_clf.values.ravel(),
                            'f1')
    
    filename = models_dir + '/clf_sgd.sav'
    pickle.dump(clf_sgd, open(filename, 'wb'))
```

```python
model_tests(clf_sgd, X_test_clf,  y_test_clf.values.ravel())
```

```python
if 'clf_KN.sav' in os.listdir(models_dir):
    clf_KN = pickle.load(open(models_dir + '/clf_KN.sav', 'rb'))
else:
    clf_KN = prepare_model(KNeighborsClassifier(), 
                                      {'n_neighbors':[3, 4, 5 , 6], 
                                       'weights':['uniform', 'distance'],
                                       'algorithm':['auto', 'ball_tree', 'kd_tree']},
                                      X_train_clf,
                                      y_train_clf.values.ravel(),
                                      'f1')
    
    filename = models_dir + '/clf_KN.sav'
    pickle.dump(clf_KN, open(filename, 'wb'))
```

```python
model_tests(clf_KN, X_test_clf,  y_test_clf.values.ravel())
```

```python
if 'clf_random_forest.sav' in os.listdir(models_dir):
    clf_random_forest = pickle.load(open(models_dir + '/clf_random_forest.sav', 'rb'))
else:
    clf_random_forest = prepare_model(RandomForestClassifier(), 
                                      {'max_depth':[5, 6, 7], 
                                       'n_estimators':[100, 200],
                                       'random_state':[12], 
                                       'criterion':['gini','entropy']},
                                      X_train_clf,
                                      y_train_clf.values.ravel(),
                                      'f1')
    
    filename = models_dir + '/clf_random_forest.sav'
    pickle.dump(clf_random_forest, open(filename, 'wb'))
```

```python
model_tests(clf_random_forest, X_test_clf,  y_test_clf.values.ravel())
```

```python
if 'clf_gradient_boosting.sav' in os.listdir(models_dir):
    clf_gradient_boosting = pickle.load(open(models_dir + '/clf_gradient_boosting.sav', 'rb'))
else:
    clf_gradient_boosting = prepare_model(GradientBoostingClassifier(), 
                                          {'learning_rate':[0.1, 0.05],
                                           'max_depth':[None, 10], 
                                           'n_estimators':[100, 200],
                                           'random_state':[12]},
                                          X_train_clf,
                                          y_train_clf.values.ravel(),
                                          'f1')
    
    filename = models_dir + '/clf_gradient_boosting.sav'
    pickle.dump(clf_gradient_boosting, open(filename, 'wb'))
```

```python
model_tests(clf_gradient_boosting, X_test_clf,  y_test_clf.values.ravel())
```

### Voting Classifier

```python
if 'voting_clf.sav' in os.listdir(models_dir):
    voting_clf = pickle.load(open(models_dir + '/voting_clf.sav', 'rb'))
else:
    voting_clf = VotingClassifier(estimators=[
        ('p1', clf_logistic_regression[2]), 
        ('p2', clf_random_forest[2]),
        ('p3', clf_gradient_boosting[2])
    ], voting='soft')
    
    voting_clf.fit( X_train_clf, y_train_clf.values.ravel())

    filename = models_dir + '/voting_clf.sav'
    pickle.dump(voting_clf, open(filename, 'wb'))
    
model_tests(voting_clf, X_test_clf,  y_test_clf.values.ravel(), from_grid=False)
```
